```python

def print_solution(board):
    for row in board:
        for i in row:
            if i==1:
                print(' Q ',end="")
            else:
                print(' . ',end="")
        print("\n")

def is_safe(board,row,col,n):
    #horizontal
    for j in range(n):
        if board[row][j] == 1:
            return False

    #vertical
    for j in range(n):
        if board[j][col] == 1:
            return False

    #left diagonal
    for i,j in zip(range(row,-1,-1),range(col,-1,-1)):
        if(board[i][j]==1):
            return False
    
    #right diagonal
    for i,j in zip(range(row,-1,-1),range(col,n,1)):
        if board[i][j]==1:
            return False
    return True

def solve_nq_util(board,row,n):
    if row == n:
        return True
    for j in range(n):
        if is_safe(board,row,j,n):
            board[row][j]=1
            if solve_nq_util(board,row+1,n):
                return True
            board[row][j]=0
    return False

def solve_nq(n):
    board = [[0]*n for i in range(n)]   # board
    if solve_nq_util(board,0,n):
        print_solution(board)
    else:
        print("solution does not exist")
        
solve_nq(8)
    
```
