```python
def water_jug_dfs(x,y,target):
    visited=set()
    path=[]

    def dfs(jug1,jug2):

        if((jug1,jug2) in visited):
            return False
        
        visited.add((jug1,jug2))   #set of tuples
        path.append((jug1,jug2))

        #if target found
        if(target==jug1 or target==jug2):
            return True

        #jugs fill condition
        if dfs(x,jug2): return True    #jug1 fill condition
        if dfs(jug1,y): return True    #jug2 fill condition
        #jugs empty conditon
        if dfs(0,jug2):return True   #jug1 empty
        if dfs(jug1,0):return True   #jug2 empty
        #pouring condtion
        if dfs(max(0,jug1-(y-jug2)),min(y,jug1+jug2)): return True   #pour A->B
        if dfs(min(x,jug1+jug2),max(0,jug2-(x-jug1))): return True   #pour B->A

        path.pop()
        return False

    dfs(0,0)
    return path

solution=water_jug_dfs(4,3,2)
if(solution):
    # for steps in solution:
    #     print(steps)
    print(*solution,sep="\n")
else:
    print("no solution")
```

```python
# water jug problem

def water_jug_dfs(x,y,target):    #x-capacity of A , y-capacity of B, target-goal state
    visited = set()
    path =[]

    def dfs(jug1,jug2):

        if (jug1,jug2) in visited:
            return False
        
        visited.add((jug1,jug2))
        path.append((jug1,jug2))

        if(jug1==target or jug2==target):
            return True
        
        #Fill condition
        if dfs(x,jug2): return True     #jug1 fill 
        if dfs(jug1,y): return True     #jug2 fill 
        #Empty condition
        if dfs(0,jug2): return True     #jug1 empty 
        if dfs(jug1,0): return True     #jug2 empty
        #pour condition
        if dfs(max(0,jug1-(y-jug2)), min(y,jug1+jug2)): return True   #pour A->B
        if dfs(min(x,jug1+jug2), max(0,jug2-(x-jug1))): return True   #pour B->A

        path.pop()
        return False
    
    dfs(0,0)
    return path

solution = water_jug_dfs(5,3,4)
if(solution):
    print(*solution, sep="\n")
else:
    print("No solution possible")

```
