```python
# Python3 program to implement Traveling Salesman Problem
# using Naive (Brute Force) Approach

from sys import maxsize
from itertools import permutations

# Function to solve TSP
def travellingSalesmanProblem(graph, s):
    V = len(graph)
    # Store all vertices apart from source
    vertex = []
    for i in range(V):
        if i != s:
            vertex.append(i)

    # Initialize minimum path cost
    min_path = maxsize

    # Store best path
    best_path = []

    # Generate all permutations
    next_permutation = permutations(vertex)

    # Traverse every permutation
    for i in next_permutation:
        
        # Current path weight
        current_pathweight = 0

        # Current path
        current_path = [s]

        # Start from source
        k = s

        # Calculate current path cost
        for j in i:
            current_pathweight += graph[k][j]
            k = j
            current_path.append(j)

        # Return back to source
        current_pathweight += graph[k][s]
        current_path.append(s)

        # Update minimum path
        if current_pathweight < min_path:
            min_path = current_pathweight
            best_path = current_path

    return min_path, best_path


# Driver Code
if __name__ == "__main__":

    # Graph represented as adjacency matrix
    graph = [
        [0, 10, 15, 20],
        [10, 0, 35, 25],
        [15, 35, 0, 30],
        [20, 25, 30, 0]
    ]

    # Starting city
    s = 0

    # Function call
    min_cost, path = travellingSalesmanProblem(graph, s)

    # Output
    print("Minimum Path Cost:", min_cost)

    print("Optimal Path:", end=" ")
    print(*path,sep = "->")
```
