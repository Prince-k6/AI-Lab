```python
import random

# NIM Game - Optimal vs Random
piles = [3, 5, 7]
turn = 0

print("NIM GAME: Optimal vs Random Strategy\n")
while sum(piles) > 0:
    print(f"Piles: {piles}")
    
    if turn == 0:  # Optimal AI
        xor = 0
        for p in piles:
            xor ^= p
        for i, p in enumerate(piles):
            if (p ^ xor) < p:
                piles[i] = p ^ xor
                print(f"Optimal AI: XOR strategy applied\n")
                break
    else:  # Random AI
        i = random.randint(0, 2)
        piles[i] -= random.randint(1, piles[i])
        print(f"Random AI: Random move\n")
    
    turn = 1 - turn

print(f"Winner: {'Optimal AI' if turn == 0 else 'Random AI'}")
```
