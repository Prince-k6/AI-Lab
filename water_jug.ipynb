# Water Jug Problem - Depth First Search (DFS) Algorithm

## Problem Statement
Given two jugs of capacity `x` and `y` liters, find a sequence of steps to measure exactly `target` liters in one of the jugs.

**Operations allowed:**
1. Fill Jug 1 completely
2. Fill Jug 2 completely
3. Empty Jug 1 completely
4. Empty Jug 2 completely
5. Pour Jug 1 into Jug 2 (until Jug 2 is full or Jug 1 is empty)
6. Pour Jug 2 into Jug 1 (until Jug 1 is full or Jug 2 is empty)

## DFS Algorithm Overview
- Uses **recursive backtracking** to explore all possible states
- Tracks visited states to avoid infinite loops
- Returns the path from initial state (0,0) to target state
- Time Complexity: O(x*y) - explores at most x*y states

```python
def water_jug_dfs(x, y, target):
    """
    Solves water jug problem using Depth First Search
    
    Parameters:
    x (int): Capacity of Jug 1
    y (int): Capacity of Jug 2
    target (int): Target amount to measure
    
    Returns:
    list: Sequence of states (jug1, jug2) from start to solution
    """
    visited = set()
    path = []
    
    def dfs(jug1, jug2):
        # Check if state already visited (cycle detection)
        if (jug1, jug2) in visited:
            return False
        
        # Mark state as visited
        visited.add((jug1, jug2))
        path.append((jug1, jug2))
        
        # Check if target found in either jug
        if target == jug1 or target == jug2:
            return True
        
        # Fill Jug 1 completely
        if dfs(x, jug2):
            return True
        
        # Fill Jug 2 completely
        if dfs(jug1, y):
            return True
        
        # Empty Jug 1 completely
        if dfs(0, jug2):
            return True
        
        # Empty Jug 2 completely
        if dfs(jug1, 0):
            return True
        
        # Pour Jug 1 into Jug 2
        # New Jug1 = max(0, jug1 - (y - jug2))  [space available in Jug2]
        # New Jug2 = min(y, jug1 + jug2)        [Jug2 filled or receives all from Jug1]
        if dfs(max(0, jug1 - (y - jug2)), min(y, jug1 + jug2)):
            return True
        
        # Pour Jug 2 into Jug 1
        # New Jug1 = min(x, jug1 + jug2)        [Jug1 filled or receives all from Jug2]
        # New Jug2 = max(0, jug2 - (x - jug1))  [space available in Jug1]
        if dfs(min(x, jug1 + jug2), max(0, jug2 - (x - jug1))):
            return True
        
        return False
    
    # Start DFS from state (0, 0)
    dfs(0, 0)
    return path
```

```python
def display_solution(x, y, target, solution):
    """
    Display the solution in a formatted way
    
    Parameters:
    x, y: Jug capacities
    target: Target amount
    solution: Path returned by water_jug_dfs
    """
    if not solution:
        print(f"❌ No solution found for target = {target}L")
        return
    
    print(f"✅ Solution found! Jug 1 capacity: {x}L, Jug 2 capacity: {y}L, Target: {target}L")
    print(f"Total steps: {len(solution) - 1}\n")
    print(f"{'Step':<6} {'Jug 1':<8} {'Jug 2':<8} {'Status':<20}")
    print("-" * 50)
    
    for i, (j1, j2) in enumerate(solution):
        status = ""
        if j1 == target:
            status = "✓ Target in Jug 1"
        elif j2 == target:
            status = "✓ Target in Jug 2"
        
        print(f"{i:<6} {j1:<8} {j2:<8} {status:<20}")
    
    print("\n")
```

## DFS Exploration Visualization

Let's track how DFS explores the state space (all visited states, not just the solution path):

```python
def water_jug_dfs_with_exploration(x, y, target):
    """
    DFS that also tracks all explored states (for visualization)
    """
    visited = set()
    path = []
    explored_order = []  # Track order of exploration
    
    def dfs(jug1, jug2, depth=0):
        if (jug1, jug2) in visited:
            return False
        
        visited.add((jug1, jug2))
        path.append((jug1, jug2))
        explored_order.append((jug1, jug2, depth))
        
        if target == jug1 or target == jug2:
            return True
        
        if dfs(x, jug2, depth+1):
            return True
        if dfs(jug1, y, depth+1):
            return True
        if dfs(0, jug2, depth+1):
            return True
        if dfs(jug1, 0, depth+1):
            return True
        if dfs(max(0, jug1 - (y - jug2)), min(y, jug1 + jug2), depth+1):
            return True
        if dfs(min(x, jug1 + jug2), max(0, jug2 - (x - jug1)), depth+1):
            return True
        
        return False
    
    dfs(0, 0)
    return path, explored_order, visited

# Example: Show how many states are explored before finding solution
path, explored, all_visited = water_jug_dfs_with_exploration(5, 3, 4)

print(f"Total states explored: {len(all_visited)}")
print(f"Solution path length: {len(path)} steps\n")
print(f"Explored states: {sorted(all_visited)}\n")
print(f"Solution path: {path}")
```

## Algorithm Analysis

### Time Complexity
- **Worst Case**: O(x × y) - explores all possible states (x+1) × (y+1)
- **Best Case**: O(1) - target achievable in first step
- **Average Case**: O(x × y)

### Space Complexity
- **Visited Set**: O(x × y) - stores all visited states
- **Recursion Stack**: O(x × y) - maximum recursion depth
- **Path**: O(x × y) - solution path can be at most length of all states

### Why DFS for Water Jug Problem?
1. **Complete**: Guaranteed to find solution if it exists
2. **Simple**: Easy to implement with recursion
3. **Memory Efficient** (compared to BFS): Uses DFS stack instead of queue
4. **Works on any state space size**: No heuristics needed

## Key Points

### Algorithm Characteristics
- **Graph-based Search**: Represents problem as state graph
- **Backtracking**: Explores one branch fully before trying others
- **Cycle Detection**: Uses visited set to avoid infinite loops
- **Recursive**: Natural recursion structure mirrors backtracking

### Important Implementation Details
1. **Visited Set**: Critical to prevent infinite loops
2. **Correct Pouring Logic**: Must accurately calculate new jug levels
3. **Base Case**: Check if target reached in either jug
4. **Order of Operations**: Different order = different paths (both valid)

### When to Use DFS vs Other Algorithms
| Algorithm | Pros | Cons |
|-----------|------|------|
| **DFS** | Simple, memory efficient | Finds longest path first |
| **BFS** | Finds shortest path | Uses more memory |
| **A*** | Optimal + efficient | Requires heuristic |
| **Greedy** | Very fast | May not find solution |

## Edge Cases & Observations

### Impossible Cases
- Target larger than both jug capacities → No solution possible
- Example: 5L jug, 3L jug, target 10L → No solution

### Special Cases
1. **Target = 0**: Always solvable (start state itself)
2. **Target = Jug capacity**: Always solvable (fill one jug)
3. **Coprime capacities**: Most likely to have solutions (e.g., 5 & 3 are coprime)
4. **Non-coprime capacities**: May have limited solutions (e.g., 4 & 6, GCD=2)
